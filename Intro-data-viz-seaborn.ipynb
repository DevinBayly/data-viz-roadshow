# Data Visualization Example

Try plotting relationship between age and strength for concrete data

_Note_: For seaborn packages, will also need 

+ numpy
+ scipy
+ pandas
+ matplotlib

But they should be installed as dependencies by default

```python
import pandas
import seaborn
```

Load in the data 

```python
concrete = pandas.read_csv("data/Concrete_Data.csv")
```

Plot amount of cement vs. compressive strength

```python
seaborn.regplot(x = "Cement", y = "Compressive_Strength", data = concrete)
```

```python
cement_plot = seaborn.regplot(x = "Cement", y = "Compressive_Strength", data = concrete)
cement_plot.set(xlabel = "Cement (kg/m3)", ylabel = "Strength (MPa)")
```

```python
seaborn.regplot(x = "Age", y = "Compressive_Strength", data = concrete)
```

By default, it will add a linear regression line and confidence intervals. This does not look like a linear relationship (the compressive strength doesn't appear to increase continuously with age). Try a polynomial relationship.

```python
seaborn.regplot(x = "Age", y = "Compressive_Strength", data = concrete, order = 2)
```

Maybe a little better, but a 4-order polynomial works best.

```python
seaborn.regplot(x = "Age", y = "Compressive_Strength", data = concrete, order = 4)
```

Update axis labels with units and save it to a file

```python
age_plot = seaborn.regplot(x = "Age", y = "Compressive_Strength", data = concrete, order = 4)
age_plot.set(xlabel = "Age (days)", ylabel = "Strength (MPa)")
age_plot.get_figure().savefig("age-strength.png")
```
