```python
flowers_df = sns.load_dataset("iris")
flowers_df.sepal_width
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width)
```

```python
plt.hist(flowers_df.sepal_width,bins=5)
```

```python
import numpy as np
# now our bins are set every .25
plt.hist(flowers_df.sepal_width,bins=np.arange(2,5,.25))
```

```python
# really depends on what your visualization's goal is but this might be important
plt.hist(flowers_df.sepal_width,bins=[1,3,4,4.5])
```

```python
# stacked hist
setosa_df = flowers_df[flowers_df.species == "setosa"]
versicolor_df = flowers_df[flowers_df.species == "versicolor"]
virginica_df = flowers_df[flowers_df.species == "virginica"]
```

```python
plt.hist(setosa_df.sepal_width,alpha=.4,bins=np.arange(2,5,.25))
plt.hist(versicolor_df.sepal_width,alpha=.4,bins=np.arange(2,5,.25))
```

```python
plt.title("Distribution of Sepal Width")
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],stacked=True,bins=np.arange(2,5,.25))
```

```python
flights_df = sns.load_dataset("flights")
flights_df
```

```python
flights_pivot=flights_df.pivot(index = "month",columns="year",values="passengers")
```

```python
flights_pivot.columns
```

```python
flights_pivot.index
```

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_pivot)
plt.xlabel("year")
```

```python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_pivot,fmt="d",annot=True,cmap="viridis")
plt.xlabel("year")

plt.xticks(rotation=45)
plt.yticks(rotation=0)
```
